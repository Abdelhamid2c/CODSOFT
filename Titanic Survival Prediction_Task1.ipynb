```python
titanic['Sex'].replace({
                        'male':'M',
                        'female':'F'
                        },inplace=True)
titanic[['Name','Sex']].sample(4)
```

```python
nb_null = titanic.isnull().sum()
nb_null
```

```python
_ = round(nb_null/titanic.isnull().count() * 100,2).sort_values(ascending=False)
_
```

```python
table_null = pd.concat([nb_null,_],axis=1,keys=["total","%"])
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('ggplot')
table_null['total'].plot(kind='hist', bins=20, title='Distribution of Null Values')
plt.gca().spines[['top', 'right']]
plt.show() 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

female = titanic[titanic['Sex'] == 'F']
male = titanic[titanic['Sex'] == 'M']

ax = sns.histplot(female,
                  x='Age',
                  bins=30,
                  hue='Survived',
                  multiple='stack',
                  ax=axes[0],
                  kde=False)

ax.legend( labels=['Not Survived', 'Survived'])
ax.set_title("Female")

ax = sns.histplot(male,
                  x='Age',
                  bins=30,
                  hue='Survived',
                  multiple='stack',
                  ax=axes[1],
                  kde=False)

ax.legend(labels=['Not Survived', 'Survived'])
ax.set_title("Male")

#plt.show()

female['Survived'].value_counts()
female['Survived'].value_counts()
```

```python
female_survived_counts = female['Survived'].value_counts()
male_survived_counts = male['Survived'].value_counts()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

axes[0].pie(female_survived_counts, labels=[ 'Survived'], autopct='%1.1f%%', startangle=90, colors=['coral'])
axes[0].set_title("Female")

axes[1].pie(male_survived_counts, labels=[ 'Not Survived'], autopct='%1.1f%%', startangle=90, colors=['lightgreen'])
axes[1].set_title("Male")

plt.show()
```
