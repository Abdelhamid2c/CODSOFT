```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.heatmap(titanic.isnull(), cbar=True, yticklabels=False)
plt.title('Distribution of Null Values')
plt.show()
```

```python
# Remplir les valeurs manquantes dans la colonne 'Age' du jeu de données d'entraînement
# en utilisant la médiane de cette colonne, et appliquer le remplacement directement dans le DataFrame.

moy_age = titanic['Age'].median()
titanic['Age'].fillna(moy_age, inplace=True)

titanic['Fare'].fillna(titanic['Fare'].median(), inplace=True)
titanic.drop(columns=['Cabin','Ticket'], axis=1, inplace=True)

titanic['Embarked'].fillna('S', inplace=True)


titanic.isnull().sum()
```

```python
survival_percentage = titanic['Survived'].sum() / len(titanic) * 100
survival_percentage = round(survival_percentage)
not_survival_percentage = round(100 - survival_percentage)

plt.figure(figsize=(4, 4))
plt.pie([survival_percentage, not_survival_percentage], labels=['Survived', 'Not Survived'], autopct='%1.4f%%', colors=['green', 'red'], startangle=90)

# Set the aspect ratio to be equal, creating a circle
plt.axis('equal')

# Add a title
plt.title('Survival Percentage')

# Show the pie chart
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import confusion_matrix
from sklearn import preprocessing
```

```python
titanic.sample(2)
```

```python
x = titanic.drop(columns=['Sex','PassengerId','Survived','Name'],axis=1)
x.replace({ 'Embarked':
           {'S':1,
            'C':2,
            'Q':3}}, inplace = True)
y=titanic.Survived
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,
                                  test_size= .1, random_state = 42, stratify=y)
```

```python
x_train.shape,x_test.shape
```

```python
x_train
```

```python
b = DummyRegressor()
b.fit(x_train,y_train)
```

```python
y_pred = b.predict(x_test)
mean_squared_error(y_test,y_pred)
```

```python
lr = LogisticRegression(random_state=42)
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_true=y_test, y_pred=lr.predict(x_test))
```
