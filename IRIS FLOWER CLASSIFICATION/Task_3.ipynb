```python
iris_data['species'] = iris_data['species'].replace({
      'Iris-setosa' : '1',
      'Iris-versicolor' : '2',
      'Iris-virginica' : '3'
})
```

```python
iris_data['species'].unique()
```

```python
iris_data['species'] = pd.to_numeric(iris_data['species'], errors='coerce')
iris_data.info()
```

```python
target_1 = iris_data[iris_data['species'] == 1]
plt.scatter(target_1['sepal_length'], target_1['sepal_width'], color='blue', label='Iris-setosa')

target_2 = iris_data[iris_data['species'] == 2]
plt.scatter(target_2['sepal_length'], target_2['sepal_width'], color='orange', label='Iris-versicolor')

plt.legend()
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')

plt.show()
```

```python
X = iris_data.drop(['species'], axis=1)
Y = iris_data['species']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.2,random_state=42)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print('Mean Squared Error : ',   mean_squared_error(Y_test, lr.predict(X_test)))
print('score : ', r2_score(Y_test,lr.predict(X_test)))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_true=Y_test, y_pred=lr.predict(X_test))
```

```python
cm
```

```python
ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax)

ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
ax.yaxis.set_ticklabels(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.show()
```
